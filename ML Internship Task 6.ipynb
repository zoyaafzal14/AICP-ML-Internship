```python
#question 6:
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Label encoding for 'Age' and 'Income Level'
label_encoder_age = LabelEncoder()
df['Age_encoded'] = label_encoder_age.fit_transform(df['Age'])

label_encoder_income = LabelEncoder()
df['Income_encoded'] = label_encoder_income.fit_transform(df['Income Level'])

# Label encoding for 'Education Level'
label_encoder_education = LabelEncoder()
df['Education_encoded'] = label_encoder_education.fit_transform(df['Education Level'])

# Label encoding for 'Device Usage'
label_encoder_device = LabelEncoder()
df['Device_encoded'] = label_encoder_device.fit_transform(df['Device Usage'])

# Select relevant features for clustering
features = ['Age_encoded', 'Income_encoded', 'Education_encoded', 'Device_encoded',
            'Likes and Reactions', 'Followed Accounts',
            'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)']

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])

# Apply K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)  # Set n_init to 10 explicitly
clusters = kmeans.fit_predict(scaled_features)

# Add cluster label to the dataframe
df['Cluster'] = clusters

# Analyzing the clusters
cluster_counts = df['Cluster'].value_counts()
print("Cluster Counts")
print(cluster_counts)

# Visualizing the clusters
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Cluster', hue='Cluster', palette='Set2')  # Remove legend=False
plt.title('Cluster Distribution')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame with 'Cluster' column indicating the cluster labels

# Numerical features
numerical_features = ['Age_encoded', 'Income_encoded', 'Education_encoded', 'Device_encoded',
                      'Likes and Reactions', 'Followed Accounts',
                      'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)']

# Categorical features
categorical_features = ['Gender', 'Location', 'Language', 'Income Level', 'Top Interests']

# Define a function to compute mean for numerical features and mode for categorical features
def compute_cluster_characteristics(df, numerical_features, categorical_features):
    cluster_characteristics = {}
    for cluster in df['Cluster'].unique():
        cluster_data = df[df['Cluster'] == cluster]
        cluster_means = cluster_data[numerical_features].mean()
        cluster_modes = cluster_data[categorical_features].mode().iloc[0]  # Mode for each categorical feature
        cluster_characteristics[cluster] = {
            'Numerical Means': cluster_means,
            'Categorical Modes': cluster_modes
        }
    return cluster_characteristics

# Compute cluster characteristics
cluster_characteristics = compute_cluster_characteristics(df, numerical_features, categorical_features)

# Print the characteristics for each cluster
for cluster, characteristics in cluster_characteristics.items():
    print(f"Cluster {cluster}:")
    print("Numerical Means:")
    print(characteristics['Numerical Means'])
    print("Categorical Modes:")
    print(characteristics['Categorical Modes'])
    print()
```

```python
# Cluster names based on characteristics
cluster_names = {
    0: "Weekend Warriors",
    1: "Engaged Professionals",
    2: "Low-Key Users",
    3: "Active Explorers",
    4: "Budget Browsers"
}

# Print the characteristics and assigned names for each cluster
for cluster, characteristics in cluster_characteristics.items():
    print(f"Cluster {cluster} – {cluster_names[cluster]}:")
    print("Numerical Means:")
    print(characteristics['Numerical Means'])
    print("Categorical Modes:")
    print(characteristics['Categorical Modes'])
    print()
```

```python
#question 9:
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Define the features and their mean values for each cluster
features = ['Age', 'Likes and Reactions', 'CTR', 'Conversion_Rates']
cluster_means = {
    "Weekend Warriors": [30, 60, 0.1, 0.15],
    "Engaged Professionals": [32, 80, 0.15, 0.2],
    "Low-Key Users": [28, 50, 0.08, 0.1],
    "Active Explorers": [29, 40, 0.12, 0.18],
    "Budget Browsers": [27, 30, 0.05, 0.08]
}

# Number of features
num_features = len(features)

# Create a function to plot radar chart
def plot_radar_chart(cluster_means, cluster_names):
    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_features, endpoint=False).tolist()

    # Make the plot close
    cluster_means = np.concatenate((cluster_means, cluster_means[:,[0]]), axis=1)
    angles += angles[:1]

    # Initialize the radar chart
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    
    # Draw one axis per feature
    plt.xticks(angles[:-1], features, color='grey', size=10)
    
    # Draw y-labels
    ax.set_rlabel_position(0)
    plt.yticks([20, 40, 60, 80], ["20", "40", "60", "80"], color="grey", size=7)
    plt.ylim(0, 100)

    # Plot data
    for i in range(len(cluster_names)):
        ax.plot(angles, cluster_means[i], linewidth=1, linestyle='solid', label=cluster_names[i])
        ax.fill(angles, cluster_means[i], 'b', alpha=0.1)

    # Add legend
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

    # Show the plot
    plt.show()

# Call the function to plot the radar chart
plot_radar_chart(np.array(list(cluster_means.values())), list(cluster_means.keys()))
```
